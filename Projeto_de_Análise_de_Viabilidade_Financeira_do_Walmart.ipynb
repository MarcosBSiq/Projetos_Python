Neste projeto o objetivo é analisar os dados disponibilizados pelo cliente Walmart, famosa multinacional estadunidense de lojas de departamento, que está buscando saber em qual das 45 lojas presentes no data frame seria mais interessante realizar a expansão de seu tamanho. #Instalação e importação das bibliotecas


```python
!pip install uszipcode 
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.pyplot as plt
import plotly.graph_objects as go
import plotly.express as px
```

#Importação da base de dados

```python
# IMPORTAÇÃO DA BASE DE DADOS

df = pd.read_csv('/content/Walmart.csv')


# PRINTANDO AS 5 PRIMEIRAS LINHAS DA BASE DE DADOS

df.head()
```

#Compreensão e tratamento dos dados


```python
# Identificando informações gerais (tipagem dos dados, colunas existentes e número de registros)

df.info()


print()

# Identificnado a existência de dados em branco

print(df.isnull().sum())

# Identificando a existência de entradas duplicadas

print(df.duplicated().sum())

# Transformando o campo Date no tipo datetime

df['Date'] = pd.to_datetime(df['Date'])

df.info()
```

```python
# Adicionando colunas para identificar qual é o feriado em cada data

# Semanas do Super Bowl

df.loc[(df['Date'] == '12-02-2010')|(df['Date'] == '11-02-2011')|(df['Date'] == '10-02-2012'),'Super_Bowl'] = True
df.loc[(df['Date'] != '12-02-2010')&(df['Date'] != '11-02-2011')&(df['Date'] != '10-02-2012'),'Super_Bowl'] = False

# Semanas do Labor Day

df.loc[(df['Date'] == '10-09-2010')|(df['Date'] == '09-09-2011')|(df['Date'] == '07-09-2012'),'Labor_Day'] = True
df.loc[(df['Date'] != '10-09-2010')&(df['Date'] != '09-09-2011')&(df['Date'] != '07-09-2012'),'Labor_Day'] = False

# Semanas do Thanksgiving
df.loc[(df['Date'] == '26-11-2010')|(df['Date'] == '25-11-2011'),'Thanksgiving'] = True
df.loc[(df['Date'] != '26-11-2010')&(df['Date'] != '25-11-2011'),'Thanksgiving'] = False

# Semanas do Christmas

df.loc[(df['Date'] == '31-12-2010')|(df['Date'] == '30-12-2011'),'Christmas'] = True
df.loc[(df['Date'] != '31-12-2010')&(df['Date'] != '30-12-2011'),'Christmas'] = False

display(df)
```

```python
# Verificando se existe algum feriado diferente dos citados acima

outro_feriado = df[(df['Holiday_Flag'] == 1) & (df['Super_Bowl'] == False) & (df['Labor_Day'] == False)&(df['Thanksgiving'] == False)&(df['Christmas'] == False)]

display(outro_feriado)
```

#Análise dos dados

```python
# Gráfico das vendas semanais por loja

fig, ax = plt.subplots(figsize=(40, 15))

plt. xlabel('Store', fontsize=15), plt. ylabel('Weekly_Sales', fontsize=15)

sns.boxplot (data = df, x = df['Store'] , y = df['Weekly_Sales']).set_title("Vendas Semanais por Loja");
```

```python
# Preço médio das vendas semanais por loja e ordenadas de forma crescente pelo faturamento semanal

media_vendas = df.groupby(['Store']).mean().reset_index().sort_values('Weekly_Sales')

display(media_vendas)

```

```python
# Gráfico da média das vendas semanais por loja

fig, ax = plt.subplots(figsize=(40, 15))

plt. xlabel('Store', fontsize=15), plt. ylabel('Weekly_Sales', fontsize=15)

sns.barplot (data = media_vendas, x = media_vendas['Store'] , y = media_vendas['Weekly_Sales']).set_title('Média de Vendas Semanais por Loja');
```

```python
# Vendas semanais acumuladas por loja e ordeandas de forma crescente pelo faturamento semanal

soma_vendas = df.groupby(['Store']).sum().reset_index().sort_values('Weekly_Sales')

display(soma_vendas)
```

```python
# Gráfico das vendas acumuladas por loja

fig, ax = plt.subplots(figsize=(40, 15))

plt. xlabel('Store', fontsize=15), plt. ylabel('Weekly_Sales', fontsize=15)

sns.barplot (data = soma_vendas, x = soma_vendas['Store'] , y = soma_vendas['Weekly_Sales']).set_title('Vendas acumuladas por Loja');
```

#Analisando vendas de acordo com feriados

```python
# Gráfico das vendas em semanas de feriado e em semanas sem feriados

fig, ax = plt.subplots(figsize=(40, 15))

plt. xlabel('Store', fontsize=15), plt. ylabel('Weekly_Sales', fontsize=15)

sns.boxplot (data = df, x = 'Store' , y = 'Weekly_Sales', hue ='Holiday_Flag' ).set_title('Soma de Vendas  por Loja');
```

```python
# Filtrando as vendas apenas pelas semanas que correspondem a feriados

soma_feriados = df[(df['Holiday_Flag'] == 1)]

soma_feriados['Date'] = pd.to_datetime(soma_feriados.Date).dt.strftime('%Y/%m/%d')

display(soma_feriados)

# Plotando o Gráfico

fig, ax = plt.subplots(figsize=(40, 15))

plt. xlabel('Date', fontsize=15), plt. ylabel('Weekly_Sales', fontsize=15)

sns.barplot (data = soma_feriados, x = 'Date' , y = 'Weekly_Sales').set_title('Soma de Vendas  por feriados');
```

#Analisando a correlação de outras variáveis com o faturamento

```python
# Identificando os níveis de correlação entre vendas e demais variáveis:

sns.set(rc={'figure.figsize':(20,15)})

ax = sns.heatmap(df.corr(), annot=True, linewidths=0.5, fmt=".2f")
```

#Comparando o desempenho de vendas da loja 20 com as demais lojas

```python
# Agrupando os dados pelas datas de acordo com a média

media_vendas_semana = df.groupby(['Date']).mean().reset_index().sort_values('Weekly_Sales')

display(media_vendas_semana)

```

```python
# Filtrando as vendas semanais apenas da loja 20

vendas_semanais_loja20 = df[['Date','Store','Weekly_Sales']]

vendas_semanais_loja20 = vendas_semanais_loja20[vendas_semanais_loja20['Store']==20]

vendas_semanais_loja20
```

```python
# Fazendo um join entre os dados de venda da loja 20 e a média de vendas de todas as lojas

df2 = pd.merge(media_vendas_semana, vendas_semanais_loja20, on=["Date","Date"], how="inner").reset_index()

df2['comparativo_media_vendas'] = 'Weekly_Sales_y'>'Weekly_Sales_x'

df2['year']= df2['Date'].dt.year

df2.sort_values('Date')
```

```python
# Gráfico de comparação das vendas da loja 20 com a média de vendas de todas as lojas ao longo dos anos

fig = go.Figure()

fig.add_trace(go.Bar(x = df2['Date'], 
                     y = df2['Weekly_Sales_x'],
                     marker_color = 'Black',                    
                     name = 'Média de Vendas Semanais de Todas as Lojas'))

fig.add_trace(go.Bar(x = df2['Date'], 
                     y = df2['Weekly_Sales_y'],
                     marker_color = 'Red',
                     name = 'Vendas Semanais da loja 20'))
```
